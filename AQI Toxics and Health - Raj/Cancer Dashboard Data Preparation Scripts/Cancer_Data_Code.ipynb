```python
import pandas as pd
```

```python
df1 = pd.read_csv("United States and Puerto Rico Cancer Statistics, 1999-2014 Incidence.txt", delimiter="\t")
```

```python
df1.head()
```

```python
def state_chng(val):
    if val == "District of Columbia":
        return "District Of Columbia"
    return val
        
```

```python
df1['States'] = df1['States'].apply(state_chng)
```

```python
df1['States'].values
```

```python
df1.to_csv("Respiratory_Cancer.csv")
```
